## ▶️ EDA

```python
# import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# supress warnings
import warnings
warnings.filterwarnings("ignore")

# display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../Datasets/Zomato/zomato_dataset_small.csv', encoding='latin-1')
```

```python
df.head()
```

```python
country_codes = pd.read_excel('../Datasets/Zomato/country_code.xlsx')
```

```python
country_codes
```

```python
df = pd.merge(df, country_codes, on='Country Code', how='left')
```

```python
df.head()
```

```python
print('Shape:', df.shape, f'| Rows: {df.shape[0]}, Columns: {df.shape[1]}')
```

```python
df = df.set_index('Restaurant ID')
```

```python
df.info()
```

```python
df.describe().T
```

Check null:

```python
df.isnull().sum()
```

```python
print('>>> All features having null values:', [feature for feature in df.columns if df[feature].isnull().sum()>0])
```

Fill missing values:

```python
df['Cuisines'] = df['Cuisines'].fillna(df['Cuisines'].mode()[0])
```

```python
df.isnull().sum()
```

```python
print('>>> All features having null values:', [features for features in df.columns if df[features].isnull().sum()>0])
```

```python
correl = df[['Longitude', 'Latitude', 'Average Cost for two', 'Price range', 'Aggregate rating', 'Votes']].corr(numeric_only=True)
correl
```

```python
sns.heatmap(correl, cmap='coolwarm', annot=True);
```

```python
all_dependent_columns = list(df.columns)
all_dependent_columns
```

```python
# All numerical columns
num_columns = [col for col in all_dependent_columns if df[col].dtype != 'O']
print(f'>> Numerical Columns (count: {len(num_columns)}): \n', num_columns)

# All categorical columns
cat_columns = [col for col in all_dependent_columns if col not in num_columns]
print(f'>> Categorical Columns (count: {len(cat_columns)}): \n', cat_columns)
```

```python
num_discrete_columns = [col for col in num_columns if len(df[col].unique())<25]
print(f'>> Numerical - Discrete features (count: {len(num_discrete_columns)}): \n', num_discrete_columns)
```

```python
num_continuous_columns = [col for col in num_columns if col not in num_discrete_columns]
print(f'>> Numerical - Continuous features (count: {len(num_continuous_columns)}): \n', num_continuous_columns)
```

```python
for col in cat_columns:
    print(f'>> Unique categories in "{col}":', len(df[col].unique()))
```

```python
plt.figure(figsize=(18,6))
sns.countplot(data=df, x='Country');
```

```python
# Top 10 countries by no. of orders
plt.figure(figsize=(10,3))
df['Country'].value_counts().head(10).plot(kind='bar');
```

```python
## Top 3 countries by % order count
plt.pie(df['Country'].value_counts().values[:3], labels=df['Country'].value_counts().index[:3], autopct='%1.2f%%');
```

```python
ratings = df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Count of ratings'}) # .reset_index() to covert Series > df
ratings
```

```python
plt.figure(figsize=(18,6))
sns.barplot(ratings, x='Aggregate rating', y='Count of ratings');
```

```python
plt.figure(figsize=(18,6))
sns.barplot(data=ratings, x="Aggregate rating", y="Count of ratings", hue='Rating color', palette=['blue','red','orange','yellow','green', 'green']);
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(ratings.groupby('Rating color').sum().sort_values(by='Count of ratings', ascending=False), x='Rating color', y='Count of ratings');
```

```python
plt.figure(figsize=(18,6))
sns.countplot(data=df, x='Rating text', hue='Has Table booking');
```

```python
plt.figure(figsize=(18,6))
sns.countplot(data=df, x='Rating text', hue='Has Online delivery');
```
